```python
import cv2
import numpy as np
import os

path1 = "./Left/"
path2 = "./Right/"
outpath = "./Label/"

files1 = os.listdir(path1)
files1.sort()

files2 = os.listdir(path2)
files2.sort()

l = len(files1)

i = 0

for file1, file2 in zip(files1, files2):
    print file1, file2
    img1 = cv2.imread(path1+file1,0)
    img1[img1>0] = 1
    img2 = cv2.imread(path2+file2,0)
    img2[img2>0] = 1
    img = img1+img2
    img[img>0] = 255
    j = file2.split("_")[-1]
    cv2.imwrite(outpath+j, img)
    
    i = i+1
    
    if i>l:
        break
```
